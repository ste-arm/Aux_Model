# Auxetic Model

Adjust the sliders to change the Plot. The variables are defined in the diagram below

```python

    d = widgets.IntSlider(
        value=10,
        min=0,
        max=100,
        step=1,
        description=r'$d_u$ [$\mu m$]:',
        disabled=False,
        continuous_update=False,
        orientation='Horizontal',
        readout=True,
        readout_format='d'
    )
    t = widgets.IntSlider(
        value=10,
        min=0,
        max=100,
        step=1,
        description=r'$t_u$ [$\mu m$]:',
        disabled=False,
        continuous_update=False,
        orientation='Horizontal',
        readout=True,
        readout_format='d'
    )
    h = widgets.IntSlider(
        value=100,
        min=0,
        max=200,
        step=1,
        description=r'$h_u$ [$\mu m$]:',
        disabled=False,
        continuous_update=False,
        orientation='Horizontal',
        readout=True,
        readout_format='d'
    )
    l = widgets.IntSlider(
        value=50,
        min=0,
        max=200,
        step=1,
        description=r'$l_u$ [$\mu m$]:',
        disabled=False,
        continuous_update=False,
        orientation='Horizontal',
        readout=True,
        readout_format='d'
    )
    thetay = widgets.IntSlider(
        value=120,
        min=90,
        max=180,
        step=1,
        description=r'$\theta_y$ [$^{\circ}$]:',
        disabled=False,
        continuous_update=False,
        orientation='Horizontal',
        readout=True,
        readout_format='d'
    )


    def CB(d,t,h,l,thetay):
        a = np.arange(-90, 91, 1).tolist()
        a = np.radians(a)
        hprime, lprime, PA, X1, X2, UCA, Cass, thetacb, X1prime, X2prime, rw, thetaw, crit = [], [], [], [], [], [], [], [], [] ,[], [], [], []
        for i in range(len(a)):
            hprime.append(h-(t/(np.tan(((np.pi/2)+a[i])/2))))
            lprime.append(l-((t/2)*((1/(np.tan(((np.pi/2)+a[i])/2)))+(np.tan(a[i])))))        
            PA.append(4*((hprime[i]*lprime[i]*np.cos(a[i]))+((np.square(lprime[i]))*np.sin(a[i])*np.cos(a[i]))))
            X1.append(2*l*np.cos(a[i]))
            X1prime.append(2*lprime[i]*np.cos(a[i]))
            X2prime.append(2*(hprime[i]+(lprime[i]*np.sin(a[i]))))
            X2.append(2*(h+(l*np.sin(a[i]))))
            UCA.append(X1[i]*X2[i])
            Cass.append(1-(PA[i]/UCA[i]))
            rw.append((UCA[i]+(X1prime[i]*d)+(X2prime[i]*d))/(UCA[i]))
            thetacb.append(np.degrees(np.arccos((Cass[i]*np.cos(np.radians(thetay)))-(1-Cass[i]))))
            thetaw.append(np.degrees(np.arccos(rw[i]*np.cos(np.radians(thetay)))))
            
            if thetay < np.arccos(-((1-Cass[i])/(rw[i]-Cass[i]))):
                crit.append(thetacb[i])
            else:
                crit.append(thetaw[i])
                

        fig, ax = plt.subplots()
        
        xmin=-90
        xmax=90
        ymin=0
        ymax=180

        ax.set_xlim([xmin,xmax])
        ax.set_ylim([ymin,ymax])

        ax.set_xlim([xmin,xmax])
        ax.set_ylim([ymin,ymax])

        x = [-90, -45, 0, 45, 90]
        values = [-90, -45, 0, 45, 90]
        plt.xticks(x,values)

        xlabel=r'Auxetic Rotation Angle, $\alpha$ [$^{\circ}$]'
        ylabel=r'Contact-angle, $\theta$ [$^{\circ}$]'

        plt.xlabel(xlabel)
        plt.ylabel(ylabel)

        
        ax.grid(which='major', color='#BBBBBB', linewidth=0.8)
        ax.grid(which='minor', color='#CCCCCC', linestyle=':', linewidth=0.5)
        ax.minorticks_on()
        
        ax.plot(np.degrees(a),thetacb,label=r'$\theta_{CB}$')
        ax.plot(np.degrees(a),thetaw,label=r'$\theta_{W}$')
#        ax.plot(np.degrees(a),crit)
        plt.legend(loc='lower right',ncol = 1) 
        plt.show()
    out = widgets.interactive_output(CB, {'d': d, 't': t, 'h': h, 'l': l, 'thetay':thetay})
    widgets.HBox([widgets.VBox([d, t, h, l,thetay]), out])
```
